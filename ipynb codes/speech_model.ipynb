```python
# ===============================================
# 🔊 Speech Emotion Detection (HuBERT + MetaData)
# ✅ Multi-Input: Audio + Gender + Intensity
# ✅ CPU Compatible | 🚀 Enhanced Accuracy | 📊 SHAP Fixed
# ===============================================

# ==== 1. Install Dependencies ====
#!pip install tensorflow tensorflow_hub librosa soundfile matplotlib scikit-learn shap torchaudio --quiet

# ==== 2. Imports ====
import os
import numpy as np
import pandas as pd
import librosa
import torch
import torchaudio
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
import shap
import joblib

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout, Concatenate, BatchNormalization
from tensorflow.keras.callbacks import ModelCheckpoint, TensorBoard, EarlyStopping, ReduceLROnPlateau
from datetime import datetime

# ==== 3. Path Setup ====
AUDIO_DIR = "/content/drive/MyDrive/hackathon/Audio_Speech_Actors_01-24"
MODEL_PATH = os.path.join(AUDIO_DIR, "final_multiinput_model_hubert123.keras")
ENCODER_PATH = os.path.join(AUDIO_DIR, "label_encoder1123.pkl")

# ==== 4. Emotion Mapping ====
emotion_map = {
    "01": "neutral",
    "02": "calm",
    "03": "happy",
    "04": "sad",
    "05": "angry",
    "06": "fearful",
    "07": "disgust",
    "08": "surprised"
}

# ==== 5. Parse Metadata ====
def parse_filename(file_name):
    parts = file_name.split("-")
    if len(parts) < 7:
        return None
    emotion_id = parts[2]
    intensity = parts[3]
    actor_id = parts[6].split(".")[0]
    gender = "female" if int(actor_id) % 2 == 0 else "male"
    return {
        "emotion": emotion_map.get(emotion_id, "unknown"),
        "intensity": "strong" if intensity == "02" else "normal",
        "gender": gender,
        "actor": int(actor_id)
    }

# ==== 6. Build Dataset from File Names ====
records = []
for root, _, files in os.walk(AUDIO_DIR):
    for file in files:
        if file.endswith(".wav"):
            meta = parse_filename(file)
            if meta and meta['emotion'] != "unknown":
                records.append({
                    "path": os.path.join(root, file),
                    **meta
                })

df = pd.DataFrame(records)
print("Sample Metadata:")
print(df.head())

# ==== 7. Load HuBERT ==== (from torchaudio)
hubert_bundle = torchaudio.pipelines.HUBERT_LARGE
hubert_model = hubert_bundle.get_model()
hubert_model.eval()

# ==== 8. Extract Audio Embeddings (with Augmentation) ====
def load_audio(file_path):
    wav, sr = librosa.load(file_path, sr=16000)
    return wav.astype(np.float32)

def extract_embedding_hubert(waveform):
    if isinstance(waveform, np.ndarray):
        waveform = waveform.astype(np.float32)
        waveform = torch.tensor(waveform)
    if waveform.dtype != torch.float32:
        waveform = waveform.to(torch.float32)
    if len(waveform.shape) == 1:
        waveform = waveform.unsqueeze(0)
    with torch.no_grad():
        features = hubert_model(waveform)[0]
        return torch.mean(features, dim=1).squeeze().numpy()

def augment_audio(audio, sr=16000):
    if audio.dtype != np.float32:
        audio = audio.astype(np.float32)
    if np.random.rand() < 0.7:
        noise = np.random.normal(0, 0.004, audio.shape)
        audio = audio + noise
    if np.random.rand() < 0.5:
        audio = librosa.effects.pitch_shift(y=audio, sr=sr, n_steps=np.random.uniform(-3, 3))
    if np.random.rand() < 0.5:
        audio = librosa.effects.time_stretch(audio, rate=np.random.uniform(0.85, 1.15))
    return np.clip(audio, -1.0, 1.0)

X_audio, gender_list, intensity_list, y_emotion = [], [], [], []

for idx, row in df.iterrows():
    try:
        waveform = load_audio(row['path'])
        if np.random.rand() < 0.6:
            waveform = augment_audio(waveform)
        emb = extract_embedding_hubert(waveform)
        X_audio.append(emb)
        gender_list.append(row['gender'])
        intensity_list.append(row['intensity'])
        y_emotion.append(row['emotion'])
    except Exception as e:
        print(f"Failed: {row['path']} - {e}")

X_audio = np.array(X_audio)

# ==== 9. Encode Categorical Inputs ====
le_emotion = LabelEncoder()
le_gender = LabelEncoder()
le_intensity = LabelEncoder()

y_encoded = le_emotion.fit_transform(y_emotion)
gender_encoded = le_gender.fit_transform(gender_list)
intensity_encoded = le_intensity.fit_transform(intensity_list)

joblib.dump(le_emotion, ENCODER_PATH)

# ==== 10. Split Dataset ====
X_train_audio, X_test_audio, gender_train, gender_test, intensity_train, intensity_test, y_train, y_test = train_test_split(
    X_audio, gender_encoded, intensity_encoded, y_encoded, test_size=0.2, stratify=y_encoded, random_state=42
)

# ==== 11. Build Multi-Input Model ====
input_audio = Input(shape=(1024,), name='audio_input')
input_gender = Input(shape=(1,), name='gender_input')
input_intensity = Input(shape=(1,), name='intensity_input')

x = Dense(2048, activation='relu')(input_audio)
x = BatchNormalization()(x)
x = Dropout(0.5)(x)
x = Dense(1024, activation='relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.5)(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.4)(x)

meta = Concatenate()([input_gender, input_intensity])
meta = Dense(128, activation='relu')(meta)
meta = Dropout(0.4)(meta)

combined = Concatenate()([x, meta])
combined = Dense(512, activation='relu')(combined)
combined = BatchNormalization()(combined)
combined = Dropout(0.5)(combined)
final = Dense(len(np.unique(y_encoded)), activation='softmax')(combined)

model = Model(inputs=[input_audio, input_gender, input_intensity], outputs=final)
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0003), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()

# ==== 12. Training ====
checkpoint = ModelCheckpoint(MODEL_PATH, monitor='val_accuracy', save_best_only=True, verbose=1)
earlystop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', patience=5, factor=0.5, verbose=1)

log_dir = "/content/logs/" + datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard = TensorBoard(log_dir=log_dir)

history = model.fit(
    [X_train_audio, gender_train, intensity_train], y_train,
    validation_data=([X_test_audio, gender_test, intensity_test], y_test),
    epochs=150,
    batch_size=32,
    callbacks=[checkpoint, tensorboard, earlystop, reduce_lr],
    verbose=2
)

# ==== 13. Evaluation ====
model.load_weights(MODEL_PATH)
y_pred = np.argmax(model.predict([X_test_audio, gender_test, intensity_test]), axis=1)

print("Classification Report:\n")
print(classification_report(y_test, y_pred, target_names=le_emotion.classes_))

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=le_emotion.classes_, yticklabels=le_emotion.classes_)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

# ==== 14. SHAP Explainability (Fixed Input Shape) ====
print("SHAP Explanation for Top 10 Test Samples...")

def shap_model(X):
    gender_dummy = np.zeros((X.shape[0], 1))
    intensity_dummy = np.zeros((X.shape[0], 1))
    return model.predict([X, gender_dummy, intensity_dummy])

X_shap_subset = X_train_audio[:100]
X_test_subset = X_test_audio[:10]

explainer = shap.KernelExplainer(shap_model, X_shap_subset)
shap_values = explainer.shap_values(X_test_subset, nsamples=100)

shap.summary_plot(shap_values, X_test_subset, feature_names=[f'feat_{i}' for i in range(X_test_audio.shape[1])])

# ==== 15. Save Final Model ====
model.save(MODEL_PATH)
print(f"✅ Multi-input HuBERT model saved to: {MODEL_PATH}")
print(f"✅ Emotion label encoder saved to: {ENCODER_PATH}")
```
