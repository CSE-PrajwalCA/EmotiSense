```python
# Complete Colab-ready code: Fine-tune EfficientNetB0 with FER2013-like folder structure,
# include data augmentation, dual-output autoencoder, Grad-CAM & LIME explainability, and save model.

import os
from pathlib import Path

import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.layers import Input, Dense, GlobalAveragePooling2D, Conv2DTranspose, Reshape
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau

# XAI imports
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
from tensorflow.keras import backend as K
from lime import lime_image
from skimage.segmentation import mark_boundaries

# # Mount Google Drive (if running in Colab)
# from google.colab import drive
# drive.mount('/content/drive')

# Paths
base_dir = Path('/content/drive/MyDrive/Hackathon/FER2013')  # adjust as needed
train_dir = base_dir / 'train'
test_dir = base_dir / 'test'
save_dir = Path('/content/drive/MyDrive/Hackathon/FER_model')
save_dir.mkdir(parents=True, exist_ok=True)

# Parameters
batch_size = 32
img_size = (128, 128)
num_classes = 7
epochs = 20

# Data Generators with Augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)

test_datagen = ImageDataGenerator(rescale=1./255)

train_gen = train_datagen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical'
)

test_gen = test_datagen.flow_from_directory(
    test_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)

# Build Dual-Output Model: Classification + Autoencoder
# 1. Encoder: EfficientNetB0 without top
base_input = Input(shape=(*img_size, 3))
base_model = EfficientNetB0(include_top=False, input_tensor=base_input, weights='imagenet')

# Classification head
a = GlobalAveragePooling2D()(base_model.output)
class_output = Dense(num_classes, activation='softmax', name='class_output')(a)

# Decoder head (simple upsampling autoencoder)
# Decoder head (upsample back to 128×128)
x = base_model.output
# 4×4 → 8×8
x = Conv2DTranspose(512, (3,3), strides=(2,2), padding='same', activation='relu')(x)
# 8×8 → 16×16
x = Conv2DTranspose(256, (3,3), strides=(2,2), padding='same', activation='relu')(x)
# 16×16 → 32×32
x = Conv2DTranspose(128, (3,3), strides=(2,2), padding='same', activation='relu')(x)
# 32×32 → 64×64  ← added
x = Conv2DTranspose(64,  (3,3), strides=(2,2), padding='same', activation='relu')(x)
# 64×64 → 128×128 ← added
x = Conv2DTranspose(32,  (3,3), strides=(2,2), padding='same', activation='relu')(x)
# Final projection to 3 channels at 128×128
dec_output = Conv2DTranspose(
    3, (3,3),
    strides=(1,1),
    padding='same',
    activation='sigmoid',
    name='dec_output'
)(x)

# # Starting from the feature map
# x = base_model.output
# x = Conv2DTranspose(256, (3,3), strides=(2,2), padding='same', activation='relu')(x)
# x = Conv2DTranspose(128, (3,3), strides=(2,2), padding='same', activation='relu')(x)
# x = Conv2DTranspose(64,  (3,3), strides=(2,2), padding='same', activation='relu')(x)
# # Project back to image size
# dec_output = Conv2DTranspose(3, (3,3), strides=(1,1), padding='same', activation='sigmoid', name='dec_output')(x)

model = Model(inputs=base_input, outputs=[class_output, dec_output])

# Compile with dual losses\
model.compile(
    optimizer=Adam(1e-4),
    loss={
        'class_output': 'categorical_crossentropy',
        'dec_output': 'mse'
    },
    loss_weights={'class_output': 1.0, 'dec_output': 0.1},
    metrics={'class_output': 'accuracy'}
)

model.summary()

# Callbacks
# checkpoint = ModelCheckpoint(
#     filepath=save_dir / 'best_model.h5',
#     monitor='val_class_output_accuracy',
#     save_best_only=True,
#     mode='max'
# )
# earlystop = EarlyStopping(monitor='val_class_output_accuracy', patience=5, restore_best_weights=True)
# reduce_lr = ReduceLROnPlateau(monitor='val_class_output_loss', factor=0.5, patience=3)


# Callbacks
checkpoint = ModelCheckpoint(
    filepath=save_dir / 'best_model.keras',
    monitor='val_class_output_accuracy',
    mode='max',              # ← specify maximize
    save_best_only=True
)
earlystop = EarlyStopping(
    monitor='val_class_output_accuracy',
    mode='max',              # ← specify maximize
    patience=5,
    restore_best_weights=True
)
reduce_lr = ReduceLROnPlateau(
    monitor='val_class_output_loss',
    mode='min',              # ← loss should be minimized
    factor=0.5,
    patience=3
)




# # Train
# history = model.fit(
#     train_gen,
#     validation_data=test_gen,
#     epochs=epochs,
#     callbacks=[checkpoint, earlystop, reduce_lr]
# )
# --- wrap your existing generators ---
def combined_generator(gen):
    while True:
        x_batch, y_batch = next(gen)
        # x_batch is the images; y_batch are the one‐hot labels for class_output
        # we feed x_batch again as the target for dec_output
        yield x_batch, {'class_output': y_batch, 'dec_output': x_batch}

train_combined = combined_generator(train_gen)
test_combined  = combined_generator(test_gen)

# --- update your .fit() call ---
history = model.fit(
    train_combined,
    steps_per_epoch=len(train_gen),           # number of batches in your train generator
    validation_data=test_combined,
    validation_steps=len(test_gen),           # number of batches in your test generator
    epochs=epochs,
    callbacks=[checkpoint, earlystop, reduce_lr]
)

# # Save final model
# model.save(save_dir / 'final_model.h5')
# And use native Keras format for saving
model.save(save_dir / 'final_model.keras')


# ------------------ EXPLAINABILITY: Grad-CAM ------------------

def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    # Create a model that maps input image to last conv layer outputs + predictions
    grad_model = Model(
        [model.inputs], [model.get_layer(last_conv_layer_name).output, model.get_layer('class_output').output]
    )
    # Compute gradient of top predicted class w.r.t. last conv layer output
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(predictions[0])
        loss = predictions[:, pred_index]
    grads = tape.gradient(loss, conv_outputs)
    # Pool gradients
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    conv_outputs = conv_outputs[0]
    heatmap = tf.reduce_sum(tf.multiply(pooled_grads, conv_outputs), axis=-1)
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()

# Example: generate Grad-CAM for one test image
for i in range (5):
    test_imgs, _ = next(test_gen)
    sample_img = test_imgs[i]
    img_array = np.expand_dims(sample_img, axis=0)
    heatmap = make_gradcam_heatmap(img_array, model, last_conv_layer_name='top_conv')
    # Display
    plt.matshow(heatmap)
    plt.title('Grad-CAM Heatmap')
    plt.show()

# Overlay on original
import cv2
heatmap = cv2.resize(heatmap, img_size)
heatmap = np.uint8(255 * heatmap)
heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)
superimposed_img = heatmap * 0.4 + np.uint8(sample_img * 255)
plt.imshow(superimposed_img.astype('uint8'))
plt.axis('off')

# ------------------ EXPLAINABILITY: LIME ------------------

explainer = lime_image.LimeImageExplainer()
explanation = explainer.explain_instance(
    np.uint8(sample_img * 255),
    classifier_fn=lambda x: model.predict(x / 255.0)[0],
    top_labels=3,
    hide_color=0,
    num_samples=1000
)
# Get mask and image for top prediction
top_pred = explanation.top_labels[0]
temp, mask = explanation.get_image_and_mask(
    label=top_pred,
    positive_only=True,
    num_features=5,
    hide_rest=False
)
plt.figure(figsize=(6,6))
plt.imshow(mark_boundaries(temp / 255.0, mask))
plt.axis('off')

print("Explainability plots generated.")
```
